```python
import pandas as pd
```

```python
df=pd.read_csv('Stock_Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

The labels 0 and 1 are the dependent variables. 0 means the stock price will decrease the next day while 1 means the stock price will increase the next day

Printing the last 5 rows

```python
df.tail()
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

Put everything before the specific date given above in training dataset and everything after the specific date given above in the testing dataset

Replace all the special characters in the dataset with blanks and renaming the columns

```python
data=train.iloc[:, 2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

list1=[i for i in range(25)]
new_Index= [str(i) for i in list1]
data.columns=new_Index
data.head(5)
```

Converting the headlines to lower case

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

Combining all the 25 records to a single sentence so that they can be converted into vectors

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines=[]
for row in range (0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

Implement count vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

Implement Bag of Words by vectorizing the words in the form of (2,2)

```python
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

Implement Random Forest Classifier

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

Do the same thing for the test dataset

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

Find out the accuracy and plot confusion matrix

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
plt.figure(figsize=(9,9))
sns.heatmap(matrix,annot=True,fmt=".3f", linewidths=.5, square=True, cmap='Pastel1');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title='Accuracy : %s' % '{0:.3%}'.format(score)
plt.title(all_sample_title, size=15);
print('Accuracy : %s' % '{0:.3%}'.format(score))
```
